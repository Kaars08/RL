```python
import numpy as np
import random
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
df = pd.read_csv(r"C:\Users\karu0\Machine Learning\fun\Ads_Optimisation.csv")
```

```python
df
```

```python
N = 10000
d = 10
ads_selected = []
no_of_selections = [0] * d
sums_of_reward = [0] * d
total_reward = 0

print(no_of_selections)
```

```python
reward = df.values[1, 2]
reward
```

```python
##UCB

for i in range(0, N):
    ad = 0
    max_upper_bound = 0
    for j in range(0, d):
        if( no_of_selections[j] > 0):
            mean = sums_of_reward[j] / no_of_selections[j]
            delta = (2 * math.log(i+1) / no_of_selections[j])
            upper_bound = mean + delta
        else: 
            upper_bound = 1e400

        if upper_bound > max_upper_bound:
            max_upper_bound = upper_bound
            ad = j
    # print(ad)
    ads_selected.append(ad)
    no_of_selections[ad] += 1
    reward = df.values[i, ad]
    sums_of_reward[ad] += reward
    total_reward += reward
```

```python
from collections import Counter

ad_counts = Counter(ads_selected)
for ad, count in ad_counts.items():
    print(f"{ad + 1} : {count} selections")
```

```python
##Epsilon Greedy
epsilon = 0.4
for i in range(0, N):
    ad = 0
    highest_reward = 0
    for j in range(0, d):
        if( no_of_selections[j] > 0):
            mean = sums_of_reward[j] / no_of_selections[j]
            # delta = (2 * math.log(i+1) / no_of_selections[j])
            # upper_bound = mean + delta
        else: 
            highest_reward = 1e400

        if np.random.choice(1) > epsilon:
            ad = j
        else:
            ad = random.randrange(1, 10)
            
    ads_selected.append(ad)
    no_of_selections[ad] += 1
    reward = df.values[i, ad]
    sums_of_reward[ad] += reward
    total_reward += reward
```
